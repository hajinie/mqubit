```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = '../Docking/docking_score.csv'
df = pd.read_excel(file_path, skiprows=1)

# Rename columns for better readability
df.columns = ['Index', 'Index2', 'Protein', '7XP5_Actual', '7XP5_Theoretical', '7XP6_Actual', '7XP6_Theoretical']
df = df.drop(['Index', 'Index2'], axis=1)

# Plotting
fig, axs = plt.subplots(2, 1, figsize=(14, 12))

# 7XP5 plot
axs[0].plot(df["Protein"], df["7XP5_Actual"], marker='o', label='7XP5 Actual', color='blue')
axs[0].plot(df["Protein"], df["7XP5_Theoretical"], marker='x', label='7XP5 Theoretical', color='red')
axs[0].set_xlabel('Protein')
axs[0].set_ylabel('Values')
axs[0].set_title('7XP5: Actual vs Theoretical')
axs[0].legend()
axs[0].grid(True)

# 7XP6 plot
axs[1].plot(df["Protein"], df["7XP6_Actual"], marker='o', label='7XP6 Actual', color='green')
axs[1].plot(df["Protein"], df["7XP6_Theoretical"], marker='x', label='7XP6 Theoretical', color='orange')
axs[1].set_xlabel('Protein')
axs[1].set_ylabel('Values')
axs[1].set_title('7XP6: Actual vs Theoretical')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
